```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_heliport = pd.read_csv('data/us-heliports-landing-facilities.csv')
del df_heliport['State_Name']
del df_heliport['Facility_Type']
del df_heliport['County_s_State_Post_Office_Code']
df_airport = pd.read_csv('data/us-general-aviation-airports.csv')
del df_airport['County_s_State_Post_Office_Code']
del df_airport['State_Name']
df_vertiport = pd.concat([df_heliport, df_airport])
mask = (df_vertiport['SHAPE_X'] <-50) & (df_vertiport['SHAPE_X'] > -140) & (df_vertiport['SHAPE_Y'] <50) & (df_vertiport['SHAPE_Y'] > 22)
df_vertiport = df_vertiport[mask]
df_vertiport_atl = df_vertiport[df_vertiport['Associated_Aeronautical_Chart']=='Atlanta']
```

```python
fig = plt.figure(figsize=(15, 8), dpi=80)
ax = fig.add_subplot(111)
ax.scatter(df_vertiport['SHAPE_X'], df_vertiport['SHAPE_Y'], alpha=0.25)
plt.grid()
# ax.set_title('EM-GMM Algorithm log Likelihood Plot')
# ax.set_xlabel('Iterations')
# ax.set_ylabel('log Likelihood')
```
