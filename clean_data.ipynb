```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_heliport = pd.read_csv('data/us-heliports-landing-facilities.csv')  
mask = (df_heliport['SHAPE_X'] <-50) & (df_heliport['SHAPE_X'] > -140) & (df_heliport['SHAPE_Y'] <50) & (df_heliport['SHAPE_Y'] > 22)
df_heliport = df_heliport[mask]
df_heliport.head()
```

```python
fig = plt.figure(figsize=(15, 8), dpi=80)
ax = fig.add_subplot(111)
ax.scatter(df_heliport['SHAPE_X'], df_heliport['SHAPE_Y'])
plt.grid()
# ax.set_title('EM-GMM Algorithm log Likelihood Plot')
# ax.set_xlabel('Iterations')
# ax.set_ylabel('log Likelihood')
```
