```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_heliport = pd.read_csv('data/us-heliports-landing-facilities.csv')  
mask = (df_heliport['SHAPE_X'] <-50) & (df_heliport['SHAPE_X'] > -140) & (df_heliport['SHAPE_Y'] <50) & (df_heliport['SHAPE_Y'] > 22)
df_heliport = df_heliport[mask]
df_heliport_atl = df_heliport[df_heliport['Associated_Aeronautical_Chart']=='Atlanta']
# df_heliport_nyc = df_heliport[df_heliport['City_Name']=='Atlanta']
```
